```python
class employee():
  empcount=0
  incnt=1.10
  def __init__(self,first,last,salary):
    self.first=first
    self.last=last
    self.salary=salary
    self.email=first.lower()+"."+last.lower()+"@shivam.com"
    employee.empcount += 1
  def fullname(self):
    return f"{self.first.title()} {self.last.title()}"
  def increment(self):
    self.salary=int(self.salary*employee.incnt)
```

```python
emp1=employee("shiVa","TiwaRi",22000)
emp2=employee("RanI","SonWane",32000)
emp3=employee("shu","snm",12909)
```

```python
class manager(employee):
  def __init__(self,first,last,salary,employees=None):
    super(). __init__(first,last,salary)
    if employees is None:
      self.employees=[]
    else:
      self.employees=employees
  def addemp(self,emp):
    if emp not in self.employees:
      self.employees.append(emp)
  def removeemp(self,emp):
    if emp in self.employees:
      self.employees.remove(emp)
  def print_emps(self):
    for emp in self.employees:
      print("-->",emp.fullname())
```
