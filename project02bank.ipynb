```python
class user():
  def __init__(self,name,age,gender):
    self.name=name
    self.age=age
    self.gender=gender
  def showdetail(self):
    print("Personal detail:-")
    print(f"Name :- {self.name.title()}")
    print(f"Age :- {self.age}")
    print(f"Gender :- {self.gender.title()}")
```

```python
shiva=user("shiva Patil",21,"male")
```

```python
shiva.showdetail()
shiva.age
```
